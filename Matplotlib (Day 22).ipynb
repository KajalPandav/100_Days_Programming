```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
x=np.array([2,3,4,5,7])
y=np.array([10,25,40,150,250])
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
x=np.array([2,3,4,5,7])
y=np.array([10,25,40,150,250])
plt.plot(x,y)
plt.grid(color='r')
plt.show()
```

```python
x=np.array([2,3,4,5,7])
y=np.array([10,25,40,150,250])
plt.plot(x,y)
plt.grid(color='r',axis='x')
plt.show()
```

```python
x=np.array([2,3,4,5,7])
y=np.array([10,25,40,150,250])
plt.plot(x,y)
plt.grid(color='r',axis='y')
plt.show()
```

```python
x=np.array([2,3,4,5,7])
y=np.array([10,25,40,150,250])
plt.plot(x,y)
plt.grid(color='r',ls='--',linewidth='1')
plt.show()
```

```python
x=np.array([2,3,4,5,7])
y=np.array([10,25,40,150,250])
plt.plot(x,y)
plt.grid(color='r',ls='--',linewidth='2')
plt.show()
```

```python
x=np.array([2,3,4,5,7])
y=np.array([10,25,40,150,250])
plt.plot(x,y)
plt.grid(color='g',ls='-.',linewidth='2')
plt.title("IPL 2022")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
x=np.array([2,3,4,5,7])
y=np.array([20,35,50,120,250])
plt.subplot(1,2,1)
plt.plot(x,y,marker='o',mec='b',color='black')
plt.grid(color='y',ls='--',linewidth='1')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("IPL 2021")


x=np.array([2,3,4,5,7])
y=np.array([10,25,40,150,250])
plt.subplot(1,2,2)
plt.plot(x,y,marker='P',mec='r',color='hotpink')
plt.grid(color='r',ls='--',linewidth='1')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("IPL 2022")

plt.show()
```

```python
x=np.array([2,3,4,5,7])
y=np.array([20,35,50,120,250])
plt.subplot(2,2,1)
plt.plot(x,y,marker='o',mec='b',mfc='white',color='black')
plt.grid(color='r',ls='--',linewidth='1')
plt.title("ipl 2012")


x=np.array([2,3,4,5,7])
y=np.array([10,25,40,150,250])
plt.subplot(2,2,2)
plt.scatter(x,y)
plt.grid(color='r',ls='--',linewidth='1')
plt.title("ipl 2015")


x=np.array([2,3,4,5,7])
y=np.array([20,35,50,120,250])
plt.subplot(2,2,3)
plt.plot(x,y,marker='o',mec='b',color='black')
plt.grid(color='r',ls='--',linewidth='1')
plt.title("ipl 2010")


x=np.array([2,3,4,5,7])
y=np.array([10,25,40,150,250])
plt.subplot(2,2,4)
plt.bar(x,y)
plt.grid(color='r',ls='--',linewidth='1')
plt.title("ipl 2018")
```
