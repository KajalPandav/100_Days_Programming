# Replace all columns null values with the particular column mean

```python
min_T = weather["MinTemp"].mean()
min_T
```

```python
max_T = weather["MaxTemp"].mean()
max_T
```

```python
rain_fall = weather["Rainfall"].mean()
rain_fall
```

```python
evapo = weather["Evaporation"].mean()
evapo
```

```python
sun_shine = weather["Sunshine"].mean()
sun_shine
```

```python
wind_gustspeed = weather["WindGustSpeed"].mean()
wind_gustspeed
```

```python
humidity_9am = weather["Humidity9am"].mean()
humidity_9am
```

```python
humidity_3pm = weather["Humidity3pm"].mean()
humidity_3pm
```

```python
press_9am = weather["Pressure9am"].mean()
press_9am
```

```python
cloud_9am = weather["Cloud9am"].mean()
cloud_9am
```

```python
cloud_3pm = weather["Cloud3pm"].mean()
cloud_3pm
```

```python
temp_9am = weather["Temp9am"].mean()
temp_9am
```

```python
temp_3pm = weather["Temp3pm"].mean()
temp_3pm
```

```python
weather["MinTemp"].fillna(min_T,inplace=True)
weather["MaxTemp"].fillna(max_T,inplace=True)
weather["Rainfall"].fillna(rain_fall,inplace=True)
weather["Evaporation"].fillna(evapo,inplace=True)
weather["Sunshine"].fillna(sun_shine,inplace=True)
weather["WindGustSpeed"].fillna(wind_gustspeed,inplace=True)
weather["Humidity9am"].fillna(humidity_9am,inplace=True)
weather["Humidity3pm"].fillna(humidity_3pm,inplace=True)
weather["Pressure9am"].fillna(press_9am,inplace=True)
weather["Cloud9am"].fillna(cloud_9am,inplace=True)
weather["Cloud3pm"].fillna(cloud_3pm,inplace=True)
weather["Temp9am"].fillna(temp_9am,inplace=True)
weather["Temp3pm"].fillna(temp_3pm,inplace=True)
```
