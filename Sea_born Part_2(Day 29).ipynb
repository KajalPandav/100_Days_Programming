```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
sns.set(style='darkgrid')
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='sex',style='smoker',size='sex')
```

```python
sns.relplot(x='time',y='smoker',data=tips,hue='sex')
```

```python
sns.catplot(x='day',y='size',data=tips,kind='strip')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='strip')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='strip',hue='sex')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm',hue='sex')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='strip',hue='sex',col='smoker')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='strip',hue='sex',col='time')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='strip',hue='sex',col='size',col_wrap=3)
```
