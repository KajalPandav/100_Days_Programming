# LabelEncoding (Convert data string to numeric)

```python
from sklearn.preprocessing import LabelEncoder
L1 = LabelEncoder()
churn["Geography"]=L1.fit_transform(churn["Geography"])
churn["Gender"]=L1.fit_transform(churn["Gender"])
churn["Surname"]=L1.fit_transform(churn["Surname"])
churn
```

# Data slicing (dependent & independent variables)

```python
x = churn.iloc[:,0:13].values
x
```

```python
y = churn.iloc[:,-1].values
y
```

# Model train_test_split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train.shape,X_test.shape,y_test.shape,y_train.shape)
```

# Algorithm define

```python
ann = tf.keras.models.Sequential()
```

# input layer

```python
ann.add(tf.keras.layers.Dense(units=5, activation='relu'))
```

# first hidden layer

```python
ann.add(tf.keras.layers.Dense(units=5, activation='relu'))
```

# second hidden layer

```python
ann.add(tf.keras.layers.Dense(units=5, activation='relu'))
```

# output layer

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

# compilation

```python
ann.compile(optimizer ='adam', loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(X_train,y_train, batch_size=32, epochs=100)
```

# prediction

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# accuracy score

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
ac = accuracy_score(y_pred,y_test)*100
cm = confusion_matrix(y_pred,y_test)
cr = classification_report(y_pred,y_test)
```
