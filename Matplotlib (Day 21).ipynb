```python
x=np.array([2,5,9,10,12])
y=np.array([20,40,90,30,180])

plt.plot(x,y,color='black',ms='15',marker='*',mfc='y',mec='r',linestyle='-.')
plt.grid()
plt.show()
```

```python
x=np.array([2,5,9,10,12])
y=np.array([20,40,90,30,180])

x1=np.array([2,6,9,10,12])
y1=np.array([10,25,40,75,280])

plt.plot(x,y,color='black',ms='15',marker='*',mfc='y',mec='r',linestyle='-.')
plt.plot(x1,y1,color='blue',ms='15',marker='*',mfc='y',mec='r',linestyle='--')
plt.grid()
plt.show()
```
