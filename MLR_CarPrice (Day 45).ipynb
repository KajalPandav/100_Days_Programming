1. Linear Regression:
   Types Of Linear Regression is
   
   a) Simple Linear Regression
   b) Multiple Linear Regression
   
   a)Simple Linear Regression:
            If you have One dependent and One independent variable is called Simple Linear Regression.
            
   b)Multiple Linear Regression:
            If you have one dependent and multiple independent variables is calles Multiple Linear Regression.

# Label Encoding (Convert data string to numeric)

```python
from sklearn.preprocessing import LabelEncoder
L1 = LabelEncoder()
car["CarName"]=L1.fit_transform(car["CarName"])
car["fueltype"]=L1.fit_transform(car["fueltype"])
car["aspiration"]=L1.fit_transform(car["aspiration"])
car["doornumber"]=L1.fit_transform(car["doornumber"])
car["carbody"]=L1.fit_transform(car["carbody"])
car["drivewheel"]=L1.fit_transform(car["drivewheel"])
car["enginelocation"]=L1.fit_transform(car["enginelocation"])
car["enginetype"]=L1.fit_transform(car["enginetype"])
car["cylindernumber"]=L1.fit_transform(car["cylindernumber"])
car["fuelsystem"]=L1.fit_transform(car["fuelsystem"])
```

```python
x = car.iloc[:,0:25]
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```
