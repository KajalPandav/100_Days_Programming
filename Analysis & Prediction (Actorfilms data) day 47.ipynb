# B) predict top 10 best rating in 2020 actor movies (Use Regression)

```python
films
```

```python
from sklearn.preprocessing import LabelEncoder
L1=LabelEncoder()
films["Actor"]=L1.fit_transform(films["Actor"])
films["ActorID"]=L1.fit_transform(films["ActorID"])
films["Film"]=L1.fit_transform(films["Film"])
films["FilmID"]=L1.fit_transform(films["FilmID"])
films
```

```python
best_rating=films.loc[(films["Year"]==2020)]
best_rating
```

```python
X=best_rating[['Actor','Year']]
X
```

```python
y=best_rating[['Votes']]
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=50)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred= regressor.predict([[5527,2020]])
y_pred
```
