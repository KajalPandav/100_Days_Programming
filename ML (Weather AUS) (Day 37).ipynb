# Replace Null values using fillna function

```python
weather["MinTemp"].fillna(min_T,inplace=True)
weather["MaxTemp"].fillna(max_T,inplace=True)
weather["Rainfall"].fillna(rain_fall,inplace=True)
weather["Evaporation"].fillna(evapo,inplace=True)
weather["Sunshine"].fillna(sun_shine,inplace=True)
weather["WindGustSpeed"].fillna(wind_gustspeed,inplace=True)
weather["Humidity9am"].fillna(humidity_9am,inplace=True)
weather["Humidity3pm"].fillna(humidity_3pm,inplace=True)
weather["Pressure9am"].fillna(press_9am,inplace=True)
weather["Cloud9am"].fillna(cloud_9am,inplace=True)
weather["Cloud3pm"].fillna(cloud_3pm,inplace=True)
weather["Temp9am"].fillna(temp_9am,inplace=True)
weather["Temp3pm"].fillna(temp_3pm,inplace=True)
weather["WindGustDir"].fillna("EE",inplace=True)
weather["WindDir9am"].fillna("AF",inplace=True)
weather["WindDir3pm"].fillna("SSS",inplace=True)
weather["WindSpeed9am"].fillna(200,inplace=True)
weather["WindSpeed3pm"].fillna(300,inplace=True)
weather["Pressure3pm"].fillna(100,inplace=True)
weather["RainToday"].fillna("EE",inplace=True)
weather["RainTomorrow"].fillna("EE",inplace=True)
```

```python
weather
```

# Label Encoding (Convert the data string to numeric)

```python
from sklearn.preprocessing import LabelEncoder
L1 = LabelEncoder()
weather["Location"] = L1.fit_transform(weather["Location"])
weather["WindGustDir"] = L1.fit_transform(weather["WindGustDir"])
weather["WindDir9am"] = L1.fit_transform(weather["WindDir9am"])
weather["WindDir3pm"] = L1.fit_transform(weather["WindDir3pm"])
weather["RainTomorrow"] = L1.fit_transform(weather["RainTomorrow"])
```

```python
weather_1=weather.drop(columns=["Location","Date","RainToday"])
weather_1
```

```python
weather_1["RainTomorrow"].unique()
```

# Data slicing into dependent & independent variables

```python
x = weather_1.iloc[:,0:19]
x
```

```python
y = weather_1.iloc[:,-1]
y
```

# Model train_test_split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.4, random_state=0)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

# Algorithm define

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
classifier.fit(x_train,y_train)
```

# Prediction

```python
y_pred = classifier.predict(x_test)
y_pred
```

# Accuracy_score,Confusion_matrix,Classification_report

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
ac = accuracy_score(y_pred,y_test)*100
cm = confusion_matrix(y_pred,y_test)
cr = classification_report(y_pred,y_test)
print(ac)
print("************************************")

print(cm)
print("************************************")

print(cr)
```
