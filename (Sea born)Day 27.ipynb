```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
sns.set(style="darkgrid")
```

```python
tips = sns.load_dataset("tips")
tips
```

# relational plot

```python
sns.relplot(x="total_bill",y="tip",data=tips)
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,hue="day",style="sex")
```

# Countplot

```python
sns.countplot(x="sex",data=tips)
```

```python
sns.countplot(x="day",data=tips)
```

```python
sns.countplot(y="smoker",data=tips)
```

# Scatter plot

```python
sns.scatterplot(x="day",y="tip",data=tips)
```

```python
sns.scatterplot(x='day', y='tip', data=tips,hue="time",marker="+",)
```

```python
sns.scatterplot(x='day', y='tip', data=tips, hue="time", style="time",)
```

# Cat plot

```python
sns.catplot(x='day', y='total_bill',data=tips,kind='box')
```

```python
sns.lineplot( x = "day",y = "total_bill",data = tips)
```

```python
sns.lineplot( x = "total_bill",y = "tip",size = "day",hue = "sex",data = tips)
```
