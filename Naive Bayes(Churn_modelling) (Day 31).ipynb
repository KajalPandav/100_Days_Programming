# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
L1 = LabelEncoder()
churn["Surname"]=L1.fit_transform(churn["Surname"])
churn["Geography"]=L1.fit_transform(churn["Geography"])
churn["Gender"]=L1.fit_transform(churn["Gender"])
churn
```

# Data slicing

```python
x = churn.iloc[:,0:13].values
x
```

```python
y = churn.iloc[:,-1].values
y
```

# Model train_test_split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

# Algorithm define (Naive Bayes)

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(x_train,y_train)
```

# Prediction

```python
y_pred = classifier.predict(x_test)
y_pred
```

# Accuracy_score

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
ac = accuracy_score(y_pred,y_test)*100
cm = confusion_matrix(y_pred,y_test)
cr = classification_report(y_pred,y_test)
```
